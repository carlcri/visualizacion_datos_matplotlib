## BAR PLOTS

El gráfico de barras nos permite graficar variables categóricas, es decir, variables de texto, lo que es muy importante en el mundo de la ciencia de datos y Matplotlib ofrece ciertas características que nos facilita la vida en cuanto a graficar este tipo de variables.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
countrys = ["INDIA", "JAPAN", "MEXICO", "COLOMBIA", 'GERMANY']
population = [5000, 800, 900, 1000, 300]
```

```python
plt.bar(countrys,population)
plt.show()
```

Para el ejemplo anterior, en el eje horizontal tenemos una variable categorica. Parametros:

- width: ancho de la barra
- color: un solo color, o multiples colores para que se altenene entre si

```python
plt.bar(countrys,population, width=0.5)
plt.show()
```

```python
plt.bar(countrys,population, width=0.5, color='green')
plt.show()
```

```python
plt.bar(countrys,population, width=0.5, color=['g', 'r'])
plt.show()
```

### Xticks

Configuraciones del eje x, se usa cuado por ejemplo viene los datos de una API y no es posible modificarlos, pero dado el caso se quiera colocar en otro idioma

    plt.xticks(np.arange(5)): arange de tantos elementos como tenga el ejer x, en este caso 5.

Rotation: Rotacion de los nombres, en grados

```python
plt.bar(countrys,population, width=0.5)
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'))
plt.show()
```

```python
plt.bar(countrys,population, width=0.5)
plt.xticks(np.arange(5), ('India', 'Japon', 'Mexico', 'Colombia', 'Alemania'), rotation=45)
plt.show()
```

### barh

para barras horizontales

```python
plt.barh(countrys, population)
```

#### Ejemplo

Colocar mas contexto a las graficas anteriores

```python

population = np.array(population)

plt.barh(countrys, population, color=['indigo', 'violet'])
plt.xlabel('Population')
plt.ylabel('Countries')
plt.title('Countries by Population \n No realistic values')

plt.axvline(x=population.mean(), linestyle='dashed', label='mean')
plt.legend()
plt.show()
```

## Crear otro tipo de graficas

Existen otros tipos de gráficos que Matplotlib nos proporciona para ser mucho más certeros en nuestros análisis.

- Histograma

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.randint(1,25,200)
```

### Histograma

Parametros:
- *bins*: contenedores 
- width: ancho barra: no es muy comun modificarlo
- histtype: bar, step, entre otros

```python
plt.hist(data)
plt.show()
```

```python
plt.hist(data, bins=20)
plt.show()
```

```python
plt.hist(data, bins=20, histtype='step')
plt.show()
```

```python
plt.style.use('classic')

fig, axes = plt.subplots()
axes.hist(data, bins=20, histtype='bar', color='indigo', label='distribucion aleatoria')

axes.legend()
```

### BOXPLOT

O diagrama de bigotes. Muetra la media, el rango intercuartil.

**Parametros**

- vert: horizontal o vertical: True by default
- patch_artist: False by default. Lo rellena 
- nothc: False by default. Para que lo comprima en la mediana
- showfliers: muestra o no los outliers


```python
plt.boxplot(data, vert=False, patch_artist=True)
plt.show()
```

```python
fig, axes = plt.subplots()
axes.boxplot(data, vert=False, patch_artist=True)

fig.show()
```

```python
fig, axes = plt.subplots()
axes.boxplot(data, vert=False, patch_artist=True, notch=True)

fig.show()
```

#### Graficando Outliers y Boxplots

Los Outliers seran un dolor de cabeza si no los sabemos manejar, afortundamente se puede escogerlos o no tenerlos en cuenta para nada.

Lo primero es añadir un par de outliers al set de datos

```python
data = np.random.randint(1,25,200)

data = np.append(data, [200,202])
data[-5:]
```

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True)
plt.show()
```

Analizando el Boxplot, se observa el ignoro los outliers: muestra el rango intercuartil, el maximo y el minimo. Y los Outliers los dejo por fuera del Boxplot.

Para que no los muestre ni siquiera en la grafica, modificamos el parametro *showfliers*

```python
plt.boxplot(data, vert=False, patch_artist=True, notch=True, showfliers=False)
plt.show()
```

## Grafico de Dispersion o Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
```

```python
plt.scatter(x,y)
```

#### Ejemplo. 

Crearemos un nuevo set de datos x y y, y otro llamado area

```python
N = 100
x = np.random.rand(N)
delta_y = np.random.rand(N)
y = 2*x +delta_y

area = (30*np.random.rand(N))**2
```

```python
plt.scatter(x,y)
```

Para el siguiente caso, el tamaño de las burbujas dependera de una tercera variable, llamada *area*

```python
plt.scatter(x,y, s=area)
```

Incluiremso una cuarta varianble, llamada *colors*

```python
colors = np.random.rand(N)
```

```python
plt.scatter(x,y,s=area,c=colors)
plt.show()
```

Se puede cambiar el marker, y el nivel de transparencia, por una x, un poligono, modificando los sigueintes parametros:

- marker: 'o', 'x', 'p', '<'
- alpha: transparencia

```python
plt.scatter(x,y,s=area,c=colors, marker='p')
plt.show()
```

```python
plt.scatter(x,y,s=area,c=colors, marker='o', alpha=0.5)
plt.show()
```
