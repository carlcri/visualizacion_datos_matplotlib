## Leyendas, etiquetas, titulos, tamaño

Para dar contexto a nuestros gráficos necesitamos usar títulos, leyendas, tamaño o etiquetas, para que nuestra gráfica tenga un contexto más amplio.

- set_title: para añadir titulo
- set_xlabel: label eje x
- set_ylabel: label eje y
- figsize: tamaño de la figura
- label: añadir un label al axes
- legend: usarlo en conjunto con label

```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
x = np.linspace(0,10,21)
```

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2)
# axes
ax1.plot(x,2*x, 'c--')
ax1.plot(x,3*x, 'b--')
ax2.plot(x,x**2, 'go:')

# titulos
ax1.set_title('Funciones Lineales')
ax2.set_title('Funcion Cuadratica')

# labels
ax1.set_xlabel('X')
ax1.set_ylabel('F(X)')

ax2.set_xlabel('X')
ax2.set_ylabel('F(X)')

fig.tight_layout()
fig.show()
```

Ajustamos el tamaño de la figura en la primera linea de codigo con *figsize*

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,6))
# axes
ax1.plot(x,2*x, 'c--')
ax1.plot(x,3*x, 'b--')
ax2.plot(x,x**2, 'go:')

# titulos
ax1.set_title('Funciones Lineales')
ax2.set_title('Funcion Cuadratica')

# labels
ax1.set_xlabel('X')
ax1.set_ylabel('F(X)')

ax2.set_xlabel('X')
ax2.set_ylabel('F(X)')

fig.tight_layout()
fig.show()
```

A continuacion añadimos los labels con **label** dentro de cada axe, usarlo en conjunto con **legend**.
Recuerda puedes usar dentro de label la notacion entre $$, tal y como lo hacemos con *markdown* para escribir formulas

```python
fig, (ax1,ax2) = plt.subplots(nrows=1, ncols=2, figsize=(8,6))
# axes
ax1.plot(x,2*x, 'c--', label='2x')
ax1.plot(x,3*x, 'b--', label='3x')
ax2.plot(x,x**2, 'go:', label='$x^2$')

# legends
ax1.legend()
ax2.legend()

# titulos
ax1.set_title('Funciones Lineales')
ax2.set_title('Funcion Cuadratica')

# labels
ax1.set_xlabel('X')
ax1.set_ylabel('F(X)')

ax2.set_xlabel('X')
ax2.set_ylabel('F(X)')

fig.tight_layout()
fig.show()
```

### Como se haria, dado el caso no se quiera usar el metodo orientado a objetos

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,21)

plt.plot(x,np.cos(x), 'rD-', label = 'cos(x)')
plt.plot(x,np.sin(x), 'b--', label = 'sin(x)')
plt.title('Funciones trigonometricas')
plt.xlabel('eje $x$')
plt.ylabel('eje $y$')
plt.legend()
```

Tambien podemos indicar donde ubicar la leyenda, agregando un parametro a *legend*:
- loc=lower right, lower left, upper right, upper left

```python
plt.plot(x,np.cos(x), 'rD-', label = 'cos(x)')
plt.plot(x,np.sin(x), 'b--', label = 'sin(x)')
plt.title('Funciones trigonometricas')
plt.xlabel('eje $x$')
plt.ylabel('eje $y$')
plt.legend(loc='upper right')
```

## Colores y Estilos

Podemos personalizar mejor nuestros gráficos con diferentes colores y estilos, así, se entenderá mucho mejor nuestras gráficas. 

A continuacion se muestran todos los estilos disponibles predefenidos

```python
import numpy as np
import matplotlib.pyplot as plt

print(plt.style.available)
```

```python
x = np.linspace(0,10,21)

plt.style.use('classic')
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7,5))
```

Y ahora definamos un estilo para ver como cambia el diseño

```python
plt.style.use('bmh')

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7,5))
```

### MOdificaciones de colores tipo Matlab

Entre comillas sencilllas, primero el color, y luego el estilo de la linea, como por ejemplo: r--, b-, y.-, etc etc

```python
plt.style.use('bmh')

fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(7,5))
axes[0].plot(x,x+1, 'r--')
axes[0].plot(x,x+2, 'b-')
axes[0].plot(x,x+3, 'y.-')
axes[0].plot(x,x+4, 'go:')
```

### Modificaciones con Pyplot

Para cambiar la propiedad del color, se modifica la propiedad color, como son los *named colors*, y tambien configuracion RGB, anteponiendo un #

- alpha: nivel de transparencia de 0 a 1, siendo 1 nada transparente
- linewidth: entre mas grande, mas gruesa

linestyle
- linea continua -
- linea punteada --

named linestyles
- solid
- dotted
- dashed
- dashdot

```python
plt.style.use('dark_background')

fig, axes = plt.subplots(figsize=(7,5))
axes.plot(x,x+1, color='violet', alpha=0.3, linewidth=8, linestyle= 'dashed')
axes.plot(x,x+2, color= 'olive', linewidth=5, linestyle='--')
axes.plot(x,x+3, color='#66FF66', linestyle=':')
axes.plot(x,x+4, color='#CCFFCA', alpha=0.8, linestyle='dashdot')
```

Tambien se pueden agregar marcadores con el parameto *marker* y su tamaño con *markersize*, y su color con *markerfacecolor*

```python
fig, axes = plt.subplots(figsize=(9,7))

axes.plot(x,x+1, color='violet', alpha=0.3, linewidth=8, linestyle= 'dashed', marker='o', markersize=5, 
markerfacecolor='w')

axes.plot(x,x+2, color= 'olive', linewidth=5, linestyle='--', marker= 'x')
axes.plot(x,x+3, color='#66FF66', linestyle=':', marker='v', markersize=8)
axes.plot(x,x+4, color='#CCFFCA', alpha=0.8, marker='P')
```
