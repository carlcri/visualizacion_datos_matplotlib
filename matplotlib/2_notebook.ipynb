## Subplot

Subplot permite crear gráficos dentro de una gráfica. Esto lo hace a través de una matriz de gráficos y se puede acceder a ellos a través de índices:

subplot tiene varios parametros
- El primero numero de filas
- El segundo numero de columnas
- Y el tercero el indice entre el numero de filas y columnas definido anteriormente

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,21)
y = x**2

var_x = np.random.randint(1,5,15)
```

Creamos un subplot con una fila y dos columnas(me crea dos indice), y a continuacion el primer indice.
En el segundo renglo, que voy a graficas en el subplot que acabe de definir anteriormente

```python
# Primer Grafico
plt.subplot(1,2,1)
plt.plot(x,y, 'b--')
plt.title('funcion')

# Segundo Grafico
plt.subplot(1,2,2)
plt.plot(y,x, 'rD:')
plt.title('funcion inversa')

plt.show()
```

¡Si ves lo sencillo que es!

Ahora grafiquemos en la primera parte tanto la funcion como su inversa, y en la otra simplemente una histograma

```python
# Primera arte
plt.subplot(1,2,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'bo:')

# Segunda parte
plt.subplot(1,2,2,)
plt.hist(var_x)

plt.show()
```

#### Ejemplo

Crear un subplot de dos filas y una columna, donde se muestre la informacion anterior

```python
# Primera arte
plt.subplot(2,1,1)
plt.plot(x,y, 'r--')
plt.plot(y,x, 'bo:')

# Segunda parte
plt.subplot(2,1,2,)
plt.hist(var_x)

plt.show()
```

## Método orientado a objetos

Hay distintas maneras de hacer gráficas dentro de Matplotlib, ya vimos pyplot; sin embargo, es muy complicado personalizarla y entrar a ciertos parámetros.

### Método orientado a objetos en Matplotlib
Un objeto define una figura, esa figura es un lienzo en el cual podemos introducir diferentes gráficas(axes), de las cuales cada una posee sus propios ejes(axis).

La figura representa el todo, dentro de ella vamos a configurar gráficas las cuales contienen diferentes ejes.

Es un poco más complicado, pero en el mismo gráfico podemos personalizarlo mucho mejor.

![](../images/plt1.JPG)

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,21)
y = x**2
```

A continuacion, vamos a usar el metodo orientado a objetos, donde se crea una figura, que va a ser nuestro lienzo

![estructura](../images/estructura.JPG)

```python
fig = plt.figure()
```

Nos acaba de decir que se creo una figura. A continuacion agregamos los axes, ¿Que son los axes? pues lo que queremos dibujar. Por ahora no vamos a colocar mucha atencion en los nuemeros que se muestran. Ya lo explicaremos

```python
axes = fig.add_axes([0.1,0.1,0.5,0.9])
```

```python
axes.plot(x,y, 'rD-')
```

Ahora todo junto

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'bD:')
```

añadamos el metodo show() para que no muestre el objeto

```python
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.5,0.9])
axes.plot(x,y, 'bD:')
fig.show()
```

Hasta aca hemos creado un grafico basico

#### Ejemplo

Usando el metodo orientado a objetos, crear un grafico de la funcion y la funcion inversa. Para este caso tanto axes_1 y axes_2 son objetos totalmente distintos, solo que reposan en la misma figura.

NO confundir *axes* con *axis* son dos conceptos distintos

Lo primero es tener una estructura basica con nuestros dos objetos

```python
fig = plt.figure()
axes_1 = fig.add_axes([])
axes_2 = fig.add_axes([])

fig.show()
```

Los parametros de *add_axes* correspondes los dos primeros a la posicion, y los dos ultimos al tamaño

```python
# Definiendo los objetos
fig = plt.figure()
axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes_2 = fig.add_axes([0.1, 0.1, 0.4, 0.3])

# Incorporando el contendio de los objetos
axes_1.plot(x,y, 'r')
axes_2.plot(y,x, 'b')

fig.show()
```

Modifiquemos los parametros del axes_2 para obtener una mejor visulalizacion

```python
# Definiendo los objetos
fig = plt.figure()
axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes_2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

# Incorporando el contendio de los objetos
axes_1.plot(x,y, 'r')
axes_2.plot(y,x, 'b')

fig.show()
```

Ahora mismo, para que de un color de relleno en el grafico pequeño, para que mejore el contraste.

```python
# Definiendo los objetos
fig = plt.figure()
axes_1 = fig.add_axes([0.1, 0.1, 0.8, 0.9])
axes_2 = fig.add_axes([0.2, 0.55, 0.4, 0.3])

# Incorporando el contendio de los objetos
axes_1.plot(x,y, 'r')
axes_2.plot(y,x, 'b')

axes_2.set_facecolor('gray')

fig.show()
```

Como ves el metodo orientado a objetos ofrece muchas cualidades en comparacion a usar solo el pyplot

## SUBPLOTS

Con subplots se puede trabajar en un arreglo de gráficas a las cuales se accede a través de los índices. A diferencia de lo que vimos anteriormente con *subplot*, *subplots* con *s* esta mas orientado a objetos.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0,10,21)
y = np.sin(x)
```

```python
fig, axes = plt.subplots()
```

Solo tenenmos hasta ahora un lienzo con un grafico adentro. Para acceder al grafico simplemente haciendo:

    axes.plot()

```python
fig, axes = plt.subplots()
axes.plot(x,y, 'bo-')
```

Hasta ahi nada nuevo. Exploremos mas funcionalidades, en este caso que el grafico de adentro tenga dos columnas y una fila.

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].plot(x,y, 'bD:')
axes[1].plot(y,x, 'r--')
```

Tambien se puede definir cada axes de manera independiente

```python
figure, (ax1,ax2)= plt.subplots(ncols=2, nrows=1)
ax1.plot(x,y, 'yo-')
ax2.plot(y,x, 'b--')
```

### Ejemplo

Usar Subplots para graficar varias funciones, que tenga 2 columnas y 2 filas

```python
fig, axis = plt.subplots(2,2)
axis[0,0].plot(x, np.cos(x), 'go:')
axis[0,1].plot(x, np.sqrt(x), 'y--')
axis[1,1].plot(x, np.sin(x), 'r--')
```

#### tight_layout

Esta propiedad permite mejorar las presentacion cuando los graficos estan muy apeñuscados, como se vio anteriormente

```python
fig, axis = plt.subplots(2,2)
axis[0,0].plot(x, np.cos(x), 'go:')
axis[0,1].plot(x, np.sqrt(x), 'y--')
axis[1,0].plot(x,np.cos(2*x), 'bD:')
axis[1,1].plot(x, np.sin(x), 'r--')

fig.tight_layout()
```

#### Ejercicio

Usar *subplots* y *name colors* de tamaño 2x3

```python
fig, ((ax1,ax2,ax3), (ax4, ax5, ax6)) = plt.subplots(2,3)

ax1.plot(x,np.cos(x), color='indigo', linestyle=':' )
ax2.plot(x,np.cos(2*x), color='lime', linestyle='--')
ax3.plot(x,np.sin(x), color='gold', linestyle='-')
ax4.plot(x,np.tan(x), 'co-')
ax5.plot(x,np.sqrt(x), 'kD:')

fig.tight_layout()
```
