## PYPLOT BASICO

Pyplot es una herramienta que tiene Matplotlib para ejecutar gráficas de manera sencilla. Véamos cómo puedes lograrlo.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,21)
y = x**2 
```

Con estas sencillas instrucciones se puede hacer graficas super sencillas

```python
plt.plot(x,y)
plt.show()
```

Añadimos un parametro para cambiar el color, en un principio usaremos esta notacion 'r' 'y' 'b' y si eres fan de mattlab te parecere muy familiar 

```python
plt.plot(x,y, 'b')
```

Si adicionalmente queremos que tenga ciertos marcadores o markers, añadirmos:
- **o**  para un circulo
- **x** para equis
- **s** para rombos
- **D** para diamantes

```python
plt.plot(x,y, 'bo')
```

```python
plt.plot(x,y,'rs')
```

Si ademas le añadimos una rayita, le indicamos que queremso una continuidad
- rayita(-) para continuidad
- dos puntos(:) para una consecusion de puntos

```python
plt.plot(x,y, 'rs-')
```

```python
plt.plot(x,y, 'bD:')
```

Como te has dado cuenta al añadir pequeños caracteres de texto, se puede modificas substancialmente la forma en como se muestra la grafica, con resultados muy elegantes

#### Ejemplo

Usar los conceptos aprendidos con una grafica de histograma, luego una de tipo torta, y finalmente uan de dispersion. Para empezar se crea un arreglo de numeros aleatorios, de tamaño 15. Interpretar los resultados

```python
x = np.random.randint(1,5,15)
print(x)
plt.hist(x)
```

Observamos el 1 se repite cinco veces, el 2 tres veces y asi sucesivamente. Ahora la de tipo torta o *pie*. De la cual no encuentro ninguna interpretacion por ahora

```python
plt.pie(x)
plt.show()
```

Ahora la grafica de tipo caja, donde observamos la mediana, y los cuartiles

```python
plt.boxplot(x)
plt.show()
```

Para la grafica de scatter, volvemos a crear nuestro arreglo inicial

```python
x = np.linspace(0,10,21)
y = x**2

plt.scatter(x,y)
```

La grafica de scatter muestra la relacion entre dos variables.
