## Seaborn

Seaborn es una librería construida sobre Matplotlib, por lo que hereda todas las bondades de la misma. Fue escrita por Michael Waskom y está integrada para estructuras de Pandas por lo que está optimizada para funcionar con DataFrames.

Seaborn tiene diferentes ventajas y entre ellas encuentras principalmente:

- Tiene una gran velocidad
- Facilidad para escribir código
- Altamente customizable entre gráficas y visualizaciones


### Cuál es la estructura Básica de Seaborn

	sns.'Tipo de Grafica'(
		 data='Dataset',
		 x='Data en el eje x',
		 y='Data en el eje y',
		 hue='Variable de agrupamiento')

### Tipos de Gráficas que tiene Seaborn

Seaborn ofrece ciertas características principales para problemas específicos de visualización de datos:

- Diagramas o gráficas relacionables
- Distribución de datos
- Graficar variables categóricas

Por ejemplo:

- Relplot (relacional): scatterplot, lineplot.
- Displot (distribuciones): histplot, kdeplot, ecdfplot, rugplot.
- Catplot (categorica): stripplot, swamplot, boxplot, violinplot, pointplot, barplot.

## SET

Set lo usamos para el estilo, con el parametro *style*:
- dark
- darkgrid

Para una paleta con el parametro *palette*:
- deep
- muted
- dark

Para el tipo de fuente con *font*:
- Verdana

Para el tamaño de la fuente con *font_scale*:
- font_scale 

En el siguiente codigo,, en el eje x tendremos A,B,C,D, y en el eje Y, valores numericos

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.barplot(x=['A', 'B', 'C', 'D'], y=[1,3,2,1])
plt.show()
```

```python
sns.set(style='darkgrid')
sns.barplot(x=['A', 'B', 'C', 'D'], y=[1,3,2,1])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark')
sns.barplot(x=['A', 'B', 'C', 'D'], y=[1,3,2,1])
plt.show()
```

```python
sns.set(style='darkgrid', palette='dark', font_scale=2)
sns.barplot(x=['A', 'B', 'C', 'D'], y=[1,3,2,1])
plt.show()
```

## Parámetros más usados con Seaborn

Seaborn tiene una gran variedad de gráficos, pero también tiene ciertos parámetros para cada gráfico, vamos a ver cuáles son los más comunes:

### Parámetros más comunes al usar Seaborn

Cada uno de los gráficos utiliza ciertos parámetros que tu podrás gestionar. Empieza importando las librerías y el Dataset que vas a usar.

Puedes usar ’ ; ’ al final de la linea de código para remplazar el 'plt.show()'

- data
- $x$ y $y$
- hue: agrupamiento por alguna otra variable
- kind: algunos tipos de graficos lo tienen(kde:distribucion por densidad)
- legend: por defecto es True(quitar o agregar la leyenda)
- palette: paleta de colores(deep, muted, pastel, bright, dark, and colorblind)
- alpha: nivel de transparencia

```python
sns.set(style='darkgrid', palette='dark', font_scale=1)
sns.barplot(x=['A', 'B', 'C', 'D'], y=[1,3,2,1]);
```

Importemos ciertos datasets que vienen por defecto con seaborn, para este caso *tips* relacionado a las propinas en un restaurante. Para visualizar los datasets que vienen integrados en la libreria, ejecuta:

    sns.get_dataset_names()

```python
tips = sns.load_dataset('tips')
print(type(tips))
```

```python
tips.head()
```

A manera de ejemplo, ilustremos una distribucion de frecuencia de algunas de las variables del data set

```python
sns.displot(data=tips, x='total_bill');
```

```python
sns.displot(data=tips, x='total_bill', y='tip');
```

```python
sns.displot(data=tips, x='tip', hue='sex');
```

```python
sns.displot(data=tips, x='total_bill', y='tip', hue='sex');
```

Agreguemos el parametro *kind*

```python
sns.displot(data=tips, x='total_bill', hue='sex', kind='kde');
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', palette='dark');
```

```python
sns.displot(data=tips, x='total_bill', hue='sex', palette='dark', alpha=0.3);
```
