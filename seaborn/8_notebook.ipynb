## Heatmap

Heatmap es un tipo de gráfico enfocado a una estructura matricial. Heatmap correlaciona todas las variables numéricas del dataset

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.head(2)
```

primero, hay que llevar el data set a una estructura matrizial especial y trabajr con los mapas de calor. En el cual 1, es una correlacion directa, 0 no hay correlacion. 

Por ejemplo, entre total_bill y tip hay cierta correlacion, pues su valor es 0.67

```python
tips.corr()
```

```python
sns.heatmap(tips.corr());
```

Vamos a darle mas contexto con el parametro *annot*

```python
sns.heatmap(tips.corr(), annot=True);
```

Ahora cambiemos el mapa de colores por uno mas bacano, como es 'coolwarm'

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm');
```

Ahora separando con lineas cada uno de los cuadros, y posteriormente dandol un color especifico

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5);
```

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='g');
```

Podemos customizarlo mas:

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='g', 
    vmin=0, vmax=1);
```

Y para quitar la barra lateral...

```python
sns.heatmap(tips.corr(), annot=True, cmap='coolwarm', linewidths=5, linecolor='g', 
    vmin=0, vmax=1, cbar=False);
```
