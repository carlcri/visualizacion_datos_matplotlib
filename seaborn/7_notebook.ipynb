## Relation

Seaborn maneja los gráficos de relación entre distintas variables numéricas para hacer visualizaciones. Veamos los distintos gráficos que tiene Seaborn para trabajar este tipo de relaciones.

- scatterplot
- lineplot: un poco mas complicado, pero muy util
- relplot: el favorito del profe

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.head(2)
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip');
```

¿PUedes observar que hay una tendencia, no tan fuerte, que entre mas aumenta la cuenta, mas alta la propina?

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day');
```

Si asignamos el parametro *style* directamente a una variable categorica, como puede ser *time*, veremos que aparecen cruzecitas y circulitos

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time');
```

Es mas, la variable size en el data set, indica el numero de comensales, si asignamos el parametro size a size, entre mas grande la cruzecita o circulito, mayor numero de comensales

```python
plt.figure(figsize=(10,10))
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
```

```python
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

Se puede ir un paso mas alla, y modificar los *markers*, por ejemplo, para el Lunch un Diamante, y para el DInner un cuadrado

```python
markers = {'Lunch':'D', 'Dinner':'s'}

sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', markers=markers)
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
plt.show()
```

### Line PLot



```python
sns.lineplot(data=tips, x='total_bill', y='tip');
```

```python
sns.lineplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
plt.legend(loc='center', bbox_to_anchor=(1.12,0.5))
```

### relplot

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size');
```

Y asi mismo, si usamos el parametro kind, o tipo de grafico, podemos cambiar entre tipo muy facilmente:
- line
- scatter

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', kind='scatter');
```

Asi mismo podemos agregar una nueva variable, usando el parametro *col*

```python
sns.relplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size', col='time');
```

## Join PLot y Pair Plot

Jointplot y Pairplot son funcionalidades de Seaborn, porque con un solo comando se puede generar varios tipos de gráficos.

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
tips = sns.load_dataset('tips')
print(tips.shape)
tips.head(2)
```

```python
sns.jointplot(data=tips, x='total_bill', y='tip');
```

Muy poderoso, porque permite ver para este caso, en solo grafico, una de dispersion, y otra de distribucion. Vamos a segmentarlo por *sex*

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex');
```

Si a su vez quiero un histograma o un kde, simplemente modifico el parametro *kind*
- hist
- kde

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='kde');
```

Ahora quiero mas valor en mis graficas, mediante **marginal_ticks**, y observaras aparece una pequeña escala en las distribuciones

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True);
```

con **marginal_kws** o marginal keywords, de esta forma afectara a los histogramas, mas no a la relacion.

```python
sns.jointplot(data=tips, x='total_bill', y='tip', hue='sex', kind='hist', marginal_ticks=True,
    marginal_kws=dict(bins=25, fill=False, multiple='dodge'));
```

### Pair PLot

El toma en el eje x total_bill, tip y size. ¿Que tienen en comun estas variables? que son numericas. De esta forma cruza todas las variables numericas entre si

```python
sns.pairplot(data=tips);
```

Ahora, añadirle una variable categorica con *hue*

```python
sns.pairplot(data=tips, hue='sex');
```

Con una paleta de colores calidos y frios

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm');
```

Y con **corner** eliminamos informacion duplicada

```python
sns.pairplot(data=tips, hue='sex', palette='coolwarm', corner=True);
```
